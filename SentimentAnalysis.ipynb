```python
from textblob import TextBlob
import pandas as pd

# Sample textual data (could be tweets or reviews)
data = {'Review': [
    "I love this internship, the tasks are great!",
    "This is very difficult and frustrating.",
    "The weather is okay, nothing special.",
    "The support team was incredibly helpful and fast."
]}
df = pd.DataFrame(data)

# Preprocessing and Sentiment Logic
def get_sentiment(text):
    analysis = TextBlob(text)
    return "Positive" if analysis.sentiment.polarity > 0 else "Negative" if analysis.sentiment.polarity < 0 else "Neutral"

df['Sentiment'] = df['Review'].apply(get_sentiment)

print("Sentiment Analysis Results:")
print(df)
```
